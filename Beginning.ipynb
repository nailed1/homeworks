```python
import pandas as pd

workout = pd.read_excel("123.xlsx")
workout["Дата продажи"] = pd.to_datetime(workout["Дата продажи"])
pivot1 = workout.groupby("Группа товаров")["Стоимость"].sum().sort_values(ascending=False)
pivot1
```

```python
pivot2 = workout.groupby("Название магазина").agg({
    "Артикул": "nunique",
    "Стоимость": "sum"
}).sort_values("Стоимость", ascending=False)
pivot2
```

```python
pivot3 = workout.groupby(workout["Дата продажи"].dt.month)["Стоимость"].mean()
pivot3.index = pivot3.index.map({
    1: "Январь", 2: "Февраль", 3: "Март", 4: "Апрель", 5: "Май"
})
pivot3
```

```python
pivot4 = workout.nlargest(5, "Стоимость")[["Название", "Группа товаров", "Стоимость", "Название магазина"]]
pivot4
```

```python
pivot5 = workout.groupby("Группа товаров").agg({
    "Стоимость": ["count", "sum", "mean", "min", "max"]
})
pivot5
```
